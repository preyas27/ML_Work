```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import VotingClassifier, StackingClassifier, BaggingClassifier, AdaBoostClassifier
from array import array
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

#Data

```python
data = sklearn.datasets.load_breast_cancer()
x = data.data
y = data.target
```

```python
sc = MinMaxScaler()
x = sc.fit_transform(x)
data.data
```

```python
data_train, data_test, target_train, target_test = train_test_split(x,y, test_size = 0.15, random_state = 5)
```

#NB

```python
nbmodel = GaussianNB()
nbmodel.fit(data_train, target_train)
```

```python
target_pred = nbmodel.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred))
```

#Decision Tree

```python
dtc = DecisionTreeClassifier()
dtc.fit(data_train, target_train)
```

```python
target_pred1 = dtc.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred1))
```

#Logistic Regression

```python
lrm = LogisticRegression()
lrm.fit(data_train, target_train)
```

```python
target_pred2 = lrm.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred2))
```

```python
print("Not Accuracy:",metrics.accuracy_score(target_pred1, target_pred2))
```

#Blendering using Linear Regression

```python
lrb = LogisticRegression()
new_data = [lrm.predict(data_train), nbmodel.predict(data_train), dtc.predict(data_train)]
```

```python
new_data = np.array(new_data).T
new_data
```

```python
data_train.shape
```

```python
lrb.fit(new_data, target_train)
```

```python
test =  [lrm.predict(data_test), nbmodel.predict(data_test), dtc.predict(data_test)]
test = np.array(test).T
target_pred5 = lrb.predict(test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred5))
```

#Max Voting

```python
mvm = VotingClassifier(estimators=[('lr', lrm), ('nb', nbmodel), ('dt', dtc)])
mvm.fit(data_train, target_train)
```

```python
target_pred3 = mvm.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred3))
```

#Stacking Classsifier

```python
stc = StackingClassifier(estimators=[('lr', lrm), ('nb', nbmodel), ('dt', dtc)],final_estimator=LogisticRegression() )
stc.fit(data_train, target_train)
```

```python
target_pred6 = stc.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

#Random Forest

```python
rfc = BaggingClassifier(base_estimator = DecisionTreeClassifier(), n_estimators = 15).fit(data_train, target_train)
```

```python
target_pred6 = rfc.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

```python
rfc = BaggingClassifier(base_estimator = GaussianNB(), n_estimators = 15).fit(data_train, target_train)
target_pred6 = rfc.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

```python
rfc = BaggingClassifier(base_estimator = LogisticRegression(), n_estimators = 15).fit(data_train, target_train)
target_pred6 = rfc.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

#AdaBoost

```python
adb = AdaBoostClassifier(base_estimator = GaussianNB(), n_estimators = 15).fit(data_train, target_train)
```

```python
target_pred6 = adb.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

```python
adb = AdaBoostClassifier(base_estimator = LogisticRegression(), n_estimators = 15).fit(data_train, target_train)
target_pred6 = adb.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```

```python
adb = AdaBoostClassifier(base_estimator = DecisionTreeClassifier(ccp_alpha=0.01), n_estimators = 15).fit(data_train, target_train)
target_pred6 = adb.predict(data_test)
print("Accuracy:",metrics.accuracy_score(target_test, target_pred6))
```
